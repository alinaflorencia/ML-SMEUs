```python
import pandas as pd 
import numpy as np 
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("RAW DATASET.csv")
df.head()
```

```python
df1_copy = df.copy()
df1_copy
```

```python
stm = StemmerFactory().create_stemmer()
stop = StopWordRemoverFactory().create_stop_word_remover()
```

```python
def cleaning(word):
    word = word.lower()
    word = stm.stem(word)
    word = stop.remove(word)
    return word
```

```python
df1_copy['city_category'] = df1_copy[['City','General_Category']].agg(' '.join,axis=1)
df1_copy
```

```python
df1_copy.city_category = df1_copy.city_category.apply(cleaning)
df1_copy
```

```python
cv = CountVectorizer()
cv.fit(df1_copy['city_category'])
print("Features Name: ", list(cv.vocabulary_.keys()))
```

```python
cvt_mat= cv.transform(df1_copy['city_category']) 
cvt_mat.shape 
```

```python
cvt_mat.todense()
```

```python
similarity = cosine_similarity(cvt_mat)
similarity
```

```python
sim_df = pd.DataFrame(similarity,index=df1_copy['Name_SMEs'],columns=df1_copy['Name_SMEs'])
sim_df
```

```python
sim_df.to_csv("sim.csv", index=True)
```

```python
sim = pd.read_csv("sim.csv").set_index("Name_SMEs")
```

```python
sim
```

```python
def rec(place_name,similarity_data=sim,k=3):
    index = similarity_data.loc[:,place_name].to_numpy().argpartition(range(-1,-k,-1))
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    closest = closest.drop(place_name,errors='ignore')
    return closest
```
